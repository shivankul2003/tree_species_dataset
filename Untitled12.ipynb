```python
dataset_path = "/content/drive/MyDrive/tree species dataset"
```

```python
import os

class_folders = [f for f in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, f))]
print("Classes in dataset:", class_folders)
```

```python
import glob

for folder in class_folders:
    image_files = glob.glob(os.path.join(dataset_path, folder, '*'))
    print(f"Class '{folder}': {len(image_files)} images")
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import random

# Kisi random class folder ka ek random image dikhana
random_class = random.choice(class_folders)
image_files = glob.glob(os.path.join(dataset_path, random_class, '*'))
sample_image = random.choice(image_files)

print("Sample Image Path:", sample_image)
img = Image.open(sample_image)
plt.imshow(img)
plt.title(f"Class: {random_class}")
plt.axis('off')
plt.show()
```
